```python
import os
import pandas as pd
```

```python
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
df = school_data_complete
df.rename(columns = {'type':'schooltype'}, inplace = True)
#df.head(3)
```

```python
totalschools = len(df["school_name"].unique())
#totalschools
```

```python
def place_comma(number): 
    return ("{:,}".format(number)) 
```

```python
totalstudents = place_comma(df["student_name"].count())
#totalstudents
```

```python
totalbudget = school_data["budget"].sum()
totalbudget = "$" + place_comma(totalbudget)
#totalbudget
```

```python
averagemathscore = format(df["math_score"].sum() / totalstudents, '.2f')
#averagemathscore
```

```python
averagereadingscore = format(df["reading_score"].sum() / totalstudents, '.2f')
#averagereadingscore
```

```python
mathpass = df[df["math_score"] >= 70]
percentmathpass = format((len(mathpass) / totalstudents) * 100, '.2f')
#percentmathpass
```

```python
readpass = df[df["reading_score"] >= 70]
percentreadpass = format((len(readpass) / totalstudents) * 100, '.2f')
#percentreadpass
```

```python
overallpassingrate = format((averagemathscore + averagereadingscore) / 2, '.2f')
#overallpassingrate
```

```python
data = {'Total Schools':[totalschools], 'Total Students': [totalstudents], 'Total Budget': [totalbudget], 
        'Average Math Score':[averagemathscore], 'Average Reading Score':[averagereadingscore], 
        '% Passing Math':[percentmathpass], '% Passing Reading': [percentreadpass], 
        '% Overall Passing Rate': [overallpassingrate]} 

newdf = pd.DataFrame(data) 
```

```python
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
df = school_data_complete.set_index("school_name").sort_index()
df.rename(columns = {'type':'School Type'}, inplace = True)
df.drop(["School ID", "Student ID"], axis=1, inplace=True)
#df.head(3)
```

```python
sn = df["School Type"]
sndf = pd.DataFrame(sn).reset_index()
grouped = sndf.drop_duplicates(subset = ["school_name"], keep = 'first')
school_type = grouped.set_index("school_name")
#school_type
```

```python
mask1 = df["reading_score"] >= 70 
mask2 = df["math_score"] >= 70

pass_table = df[mask1]
pass_table2 = df[mask2]
```

```python
read_pass = pass_table.pivot_table(values="student_name", index="school_name", aggfunc='count')
#read_pass
```

```python
math_pass = pass_table2.pivot_table(values="student_name", index="school_name", aggfunc='count')
#math_pass
```

```python
school_summary = pd.DataFrame(df.pivot_table(index="school_name"))
school_summary.rename(columns = {'budget':'Total School Budget', 'math_score':'Average Math Score',
                                 'reading_score': 'Average Reading Score', 'size': 'Total Students'}, inplace = True)
pd.options.display.float_format = '{:.2f}'.format
#school_summary
```

```python
per_student_budget = pd.DataFrame({"Per Student Budget": school_summary["Total School Budget"]/school_summary["Total Students"]})
#per_student_budget
```

```python
percent_math_pass = pd.DataFrame({"% Passing Math": (math_pass["student_name"] / school_summary["Total Students"]) * 100})
#percent_math_pass
```

```python
percent_read_pass = pd.DataFrame({"% Passing Reading":(read_pass["student_name"] / school_summary["Total Students"]) * 100})
#percent_read_pass
```

```python
overall_passing = pd.DataFrame({"% Overall Passing Rate": (percent_read_pass["% Passing Reading"] + percent_math_pass["% Passing Math"]) / 2})
#overall_passing
```

```python
merged_data = pd.concat([school_summary,school_type, per_student_budget,percent_math_pass,percent_read_pass,overall_passing], axis=1)
del merged_data.index.name
merged_data = merged_data[['School Type', 'Total Students', 'Total School Budget', 'Per Student Budget', 'Average Math Score',
                            'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing Rate']]
```

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

school_spend["Spending Ranges (Per Student)"] = pd.cut(school_spend["Per Student Budget"], spending_bins, labels=group_names)
school_spend_df = school_spend.drop(["index", "School Type", "Total Students", "Total School Budget", "Per Student Budget"], axis=1)

school_spend_df = school_spend_df.groupby("Spending Ranges (Per Student)")
```

## Trends

1) Students at Charter Schools are better academic performers than those in District Schools.

2) District schools have a higher population which is most likely a factor in trend #1. When schools have a higher population,
   each classroom has a higher number of students which then leads to lack of sufficient support from one teacher to students who 
   need help or are falling behind.
   
3) The 5 bottom performing schools all have low average math scores (77 and below) with an average of only 66% passing rate. This is an 
   indication that the math program in all grade levels (9th - 12th) needs much improvement. Perhaps the schools need to hire better math    
   teachers. Or create a math tutoring after school program - the budget per student at these schools are about $645 compared to just $488 at 
   the top performing schools. It makes one wonder where that budget is focused on at the bottom performing schools.
